# Preparation

```python
import pandas as pd
import joblib

# Load the trained model pipeline
# Make sure the file name matches what you saved in the main notebook
model = joblib.load('optimized_ecommerce_model.pkl')

print("Model successfully loaded.")
```

# Create New Data

```python
# Creating a dummy dataset representing new website visitors
new_data = pd.DataFrame({
    'Administrative': [2, 0, 5],
    'Administrative_Duration': [50.0, 0.0, 120.5],
    'Informational': [0, 0, 2],
    'Informational_Duration': [0.0, 0.0, 45.0],
    'ProductRelated': [15, 5, 40],
    'ProductRelated_Duration': [400.0, 100.2, 1200.0],
    'BounceRates': [0.01, 0.20, 0.02],
    'ExitRates': [0.02, 0.20, 0.03],
    'PageValues': [25.5, 0.0, 60.2], # Critical feature
    'SpecialDay': [0.0, 0.0, 0.0],
    'Month': ['May', 'Feb', 'Nov'],
    'OperatingSystems': [2, 1, 3],
    'Browser': [2, 1, 2],
    'Region': [1, 3, 1],
    'TrafficType': [2, 1, 10],
    'VisitorType': ['Returning_Visitor', 'New_Visitor', 'Returning_Visitor'],
    'Weekend': [False, False, True]
})

print("Inference data created:")
display(new_data)
```

# Model Prediction

```python
# Performing inference
predictions = model.predict(new_data)
probabilities = model.predict_proba(new_data)[:, 1] # Probability of 'True' (Purchase)

# Combining results for easier analysis
results = new_data.copy()
results['Predicted_Purchase'] = predictions
results['Purchase_Probability'] = probabilities

print("Prediction Results:")
display(results[['Month', 'VisitorType', 'PageValues', 'Predicted_Purchase', 'Purchase_Probability']])
```


The model successfully processed raw visitor data. By utilizing the pre-built pipeline, categorical encoding and numerical scaling were handled automatically, ensuring consistent predictions.

# Conclusion for Inference

```python
# Business Logic Interpretation
for i, prob in enumerate(probabilities):
    if prob > 0.5:
        print(f"User {i+1}: High Intent - This user is likely to convert. (Probability: {prob:.2f})")
    else:
        print(f"User {i+1}: Low Intent - This user is browsing without purchase intent. (Probability: {prob:.2f})")
```


From the results, we can observe that high 'PageValues' significantly boost the purchase probability. For instance, User 3 (Nov) shows a high conversion potential, allowing for targeted marketing interventions such as limited-time offers."
